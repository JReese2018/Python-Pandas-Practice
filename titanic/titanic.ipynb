## Titanic Dataset - Filtering and Sorting

I have aquired a dataset along with some tasks in order to get more practice with pandas

The tasks are as follows:
1. Load the dataset into a pandas DataFrame.
2. Display the number of passengers in each class (Pclass).
3. Filter and display only the passengers who survived.
4. Find the average age of passengers.
5. Sort the DataFrame based on the "Fare" column in descending order.

<font size="5">1. Load the dataset into a pandas DataFrame</font>

```python
## Importing Pandas
import pandas as pd

## Reading data that was given
df = pd.read_csv('titanic_data.csv')

## Printing
df
```

<font size="5">2. Display the number of passengers in each class (Pclass).</font>

```python
## There are different ways to do this but the value_counts() method might be the easiest
## We'll have to pass in the 'Pclass' column as an argument
df.value_counts(['Pclass'])
```

<font size="5">3. Filter and display only the passengers who survived.</font>

```python
## The .loc[] function helps with filtering
df.loc[df['Survived'] == True]
```

<font size="5">4. Find the average age of passengers.</font>

```python
## You can get the average of everything with the describe() function
df.describe()
```

```python
## As we can see, the mean of the age is ~29.7 years old
## We can also easily identify the mean of a specific column
df['Age'].mean()
```

<font size="5">5. Sort the DataFrame based on the "Fare" column in descending order.</font>

```python
## The sort_values() function will get us this answer. We just need to pass in the column as an arugument and the 'ascending=False' to make it descend
## I put the head() function there to view more of the data
df.sort_values('Fare', ascending=False).head(10)
```
