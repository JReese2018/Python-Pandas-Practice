## Wine Reviews - Grouping and Aggregation

I have aquired a dataset along with some tasks in order to get more practice with pandas

The tasks are as follows:
1. Load the dataset into a pandas DataFrame.
2. Display the first 5 rows of the dataset.
3. Group the data by "country" and calculate the average wine price for each country.
4. Find the maximum and minimum wine prices for each country.
5. Calculate the total number of wines reviewed for each country.
6. Display the top 3 countries with the highest average wine prices.

<font size="5">1. Load the dataset into a pandas DataFrame</font>

```python
## Importing Pandas
import pandas as pd

## Reading the data given to us
df = pd.read_csv('wine_reveiws_data.csv')

## Printing
df
```

<font size="5">2. Display the first 5 rows of the dataset</font>

```python
## The head() function will diplay the first couple of rows
## Passing in the number as an argument will display the number of rows you want
df.head(5)
```

```python
## Let's also rename the first column to 'entry_id'
df.rename(columns={'Unnamed: 0': 'entry_id'}, inplace=True)

df.head(5)
```

<font size="5">3. Group the data by "country" and calculate the average wine price for each country.</font>

```python
## For this, I can group the countries with the groupby() function
## After that, we'll get the mean of price
df.groupby(['country']).mean(['price'])
```

```python
## For some reason, it gives me the mean of every numeric value in the df
## We can just make a new df and take out those columns
price_average = df.groupby(['country']).mean(['price'])
price_average = price_average.drop(columns=['entry_id', 'points'])

## Printing
price_average
```

<font size="5">4. Find the maximum and minimum wine prices for each country.</font>

<font size="4">Minumum</font>

```python
## Using the same groupby() function we used earlier will get us the countries again
## Then we'll use the min()
minumum_price = df.groupby(['country']).min(['price'])

## Let's take out those columns again
minumum_price = minumum_price.drop(columns=['entry_id', 'points'])

print(minumum_price)
```

<font size="4">Maximum</font>

```python
## Same process for maximum
maximum_price = df.groupby(['country']).max(['price'])

maximum_price = maximum_price.drop(columns=['entry_id', 'points'])

print(maximum_price)
```

<font size="5">5. Calculate the total number of wines reviewed for each country.</font>

```python
## Let's group by countries again
reviews_per_country = df.groupby(['country']).count()

## Dropping the unneeded columns
reviews_per_country = reviews_per_country.drop(columns=['description', 'designation', 'points', 'price', 'province', 'region_1', 'region_2', 'variety', 'winery'])

## Renaming the entry_id
reviews_per_country = reviews_per_country.rename(columns={'entry_id': 'entries'})

## Printing
reviews_per_country
```

<font size="5">6. Display the top 3 countries with the highest average wine prices.</font>

```python
## Group countries and get average wine prices
top_3_prices = df.groupby(['country']).mean(['prices'])

## Take out unneeded columns
top_3_prices = top_3_prices.drop(columns=['entry_id', 'points'])


```

```python
## Order prices from high to low
top_3_prices = top_3_prices.sort_values('price', ascending=False)
```
