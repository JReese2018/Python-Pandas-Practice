<font size="5">1. Load the dataset into a pandas DataFrame</font>

```python
## Importing Pandas
import pandas as pd

## Reading the data given to us
df = pd.read_csv('advertising.csv')

## Printing
df
```

<font size="5">2. Create a scatter plot of "TV" advertising expenditure against "Sales."</font>

```python
## To plot, we'll need to use the matplot library
import matplotlib.pyplot as plt

## Let's create the scatter plot
plt.scatter(df['TV'], df['Sales'])

## Give it a title and labels
plt.title('TV Advertising vs Sales')
plt.xlabel('TV Advertising Expenditure (in thousands)')
plt.ylabel('Sales')

## Now we can show the plot with the show() function
plt.show()
```

<font size="5">3. Plot a bar chart showing the average "Sales" for each "Radio" advertising level.</font>

```python
import matplotlib.pyplot as plt

## First we'll need to get the average sale for each radio advertising
average_sales_by_radio = df.groupby('Radio')['Sales'].mean()
radios = average_sales_by_radio.index
average_sales = average_sales_by_radio.values

## Now we can create the plot
plt.bar(radios, average_sales)
plt.title('Average Sales for Each Radio Advertising Level')
plt.xlabel('Radio Advertising Expenditure (in thousands)')
plt.ylabel('Average Sales')
plt.show()
```

<font size="5">4. Visualize the distribution of "Newspaper" advertising expenditure using a histogram.</font>

```python
## This plot tells us how many times we have spent a certain amount on Newspaper Advertsing
## For example, we can see that we've spent <= 100 about 2 times

## We just need to create the histogram
plt.hist(df['Newspaper'], bins=20, edgecolor='black')
plt.title('Distribution of Newspaper Advertising Expenditure')
plt.xlabel('Newspaper Advertising Expenditure (in thousands)')
plt.ylabel('Frequency')
plt.show()
```
