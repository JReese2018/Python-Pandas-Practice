# Matplotlib

Below are the functions and methods used in this notebook
Documentation can be found here: https://matplotlib.org/3.5.3/api/pyplot_summary.html

### Types of Graphs:
plt.plot - Line Graph
plt.bar - Bar Chart
plt.hist - Histogram
plt.pie - Pie Chart
plt.bloxplot = Box Plot
### Graph Methods
    label= - Gives your line/bar a name for the legend
    color= - Let's you color your line/bar (you can use hex codes too ex. #00FF00)
    linewidth=int - Changes the width of the line
    marker= - Let's  you add markers to each point
    makersize=int - Changes the size of the marker
    linestyle= - Changes the way the line will look
### Graph Tools:
    fontdict={'fontname': 'Chosen Font', 'fontsize' : 'Chosen Font Size} - An Argument that can change the font and size of the titles and labels
#### plt.figure(figsize=(int,int))
#### plt.show() - Get's rid of things like "[<matplotlib.lines.Line2D at 0x1e9342f7b10>]"
#### plt.title('Title Here') - Gives your graph a title   
#### plt.xlabel('Label Here') - Names your X Axis
#### plt.ylabel('Label Here') - Names your Y Axis
#### plt.xticks() and plt.yticks() - Specifies the amount of ticks you want on your graph
#### plt.savefig('Graph Name.png') - Saves graph
#### plt.legend - Creates a legend for your graph
    loc('Chosen Position') - Let's you pick where you want the graph
    bbox_to_anchor=(1.05, 1), loc='upper left' - Puts the graph on the outside of the graph

### Chart Specific Parameters/Methods
    .set_hatch('Type of Hatch') - Changes the way the bar in a bar chart looks
    autopct='%.2f %%' - Displays percent on pie chart
    explode= - Splits the pie chart

## Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic Line Graph - plt.plot() 

```python
## To make a basic line graph, you can use the plt.plot() method
x = [0,1,2,3,4]
y = [0,2,4,6,8]

## Inside of the parenthesis, you want to put the values which will either be a column of a dataframe or a variable 

plt.plot(x,y)
```

```python
## If you call plt.show(), it will get rid of the "[<matplotlib.lines.Line2D at 0x1e9342f7b10>]"
## You can also give the table a name and label the axes and change the font of the titles
plt.plot(x,y)
plt.title('Line Graph', fontdict={'fontname' : 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
## As you can see, our graph showws more ticks than needed, let's see how to take those out
plt.plot(x,y)
plt.title('Line Graph', fontdict={'fontname' : 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

## You can use the plt.xticks() and plt.yticks() methods to specify
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

plt.show()
```

```python
## Now let's add a legend and change up the coloring on this
plt.plot(x,y, label='2x', color='green', linestyle='--', linewidth=2, marker='.', markersize=10)

## You can also add more than one line to the graph
x2 = np.arange(0,4.5,0.5)
plt.plot(x2, x2**2, label='x^2', color='orange')

plt.title('Line Graph', fontdict={'fontname' : 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()

plt.show()
```

```python
## Here's a cool trick on how you can give the illusion of a line starting and not finishing yet
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, label='x^2', color='green')
plt.plot(x2[5:], x2[5:]**2, color='green', linestyle='--')


plt.title('Line Graph', fontdict={'fontname' : 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()

plt.show()
```

```python
## You can also change the size of the graph
plt.figure(figsize=(5,3))

plt.plot(x,y, label='2x', color='green', linestyle='--', linewidth=2, marker='.', markersize=10)


plt.title('Line Graph', fontdict={'fontname' : 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.legend()

## To the save the graph, you use the plt.savefig() method
#plt.savefig('mygraph.png')

plt.show()
```

## Bar Chart - plt.bar() 

```python
labels = ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels, values)

## Change the way your bars look
bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')

plt.show()
```

```python
bars = plt.bar(labels, values)

## You can also use a for loop for this
patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop())


plt.show()
```

## Real Life Examples

### Gas Prices Line Graph

```python
## First we'll import the data
gas = pd.read_csv('gas_prices.csv')

## Printing
gas.head()
```

```python
## We'll use the plt.plot() to get this data
plt.figure(figsize=(12,6))
plt.title('Gas Over The Years (in USD)')
plt.xlabel('Year')
plt.ylabel('Price')

plt.plot(gas['Year'], gas['USA'], label='USA', marker='.', color='blue')
plt.plot(gas['Year'], gas['Canada'], label='Canada', marker='.', color='red')
plt.plot(gas['Year'], gas['South Korea'], label='South Korea', marker='.', color='green')

## We'll slice the years to get every third year
plt.xticks(gas['Year'][::3])

plt.legend()
plt.show()
```

```python
## If you wanted to get every single country, you can with a for loop
## We'll also change the font size
plt.figure(figsize=(12,6))
plt.title('Gas Over The Years (in USD)', fontdict={'fontsize' : 20})
plt.xlabel('Year')
plt.ylabel('Price')

for country in gas:
    if country != 'Year':
        plt.plot(gas['Year'], gas[country], marker='.', label=country)

## We'll slice the years to get every third year
plt.xticks(gas['Year'][::3])

## bbox_to_anchor allows you to move in on the graph 
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', shadow=True)
plt.show()
```

### Fifa Data

```python
## Reading the data
fifa = pd.read_csv('fifa_data.csv')

## Printing
fifa.head(5)
```

#### Histogram

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa['Overall'], bins=bins)

plt.title('Distribution of Player Skills in FIFA 2018')
plt.ylabel('Number of Players')
plt.xlabel('Overall')

plt.xticks(bins)

plt.show()
```

#### Pie Chart

```python
## Identifying
left = fifa.loc[fifa['Preferred Foot'] == 'Left']
left_foot = len(left)
right = fifa.loc[fifa['Preferred Foot'] == 'Right']
right_foot = len(right)

## To add labels and colors, you have to make lists 
labels = ['Left', 'Right']
colors = ['orange', 'green']

## Adding the autopct parameter puts the number percent on the pie chart
plt.pie([left_foot, right_foot], labels=labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Preferrance of FIFA Players')

plt.show()
```

```python
## Let's say you wanted to get a distibution of players in different weight classes
fifa['Weight']
```

```python
## First we'll extract the numbers from that column
fifa['Weight'] = fifa['Weight'].str.extract(r'(\d+)\s*lbs')
## Then we'll turn it into an int so we can put it in a pie chart
fifa['Weight'] = fifa['Weight'].astype(float).astype('Int64')
## Now we check
fifa['Weight']
```

```python
light = len(fifa.loc[fifa['Weight'] < 125])
medium_light = len(fifa.loc[(fifa['Weight'] >= 125) & (fifa['Weight'] < 150)])
medium = len(fifa.loc[(fifa['Weight'] >= 151) & (fifa['Weight'] < 175)])
medium_heavy = len(fifa.loc[(fifa['Weight'] >= 175) & (fifa['Weight'] < 200)])
heavy = len(fifa.loc[fifa['Weight'] >= 200])

weights = [light, medium_light, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (.4,.1,0,0,.4)

plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)

plt.title('Weight Distribution of FIFA Players (in lbs)')

plt.show()
```

```python
barcelona = fifa.loc[fifa['Club'] == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa['Club'] == 'Real Madrid']['Overall']
revs = fifa.loc[fifa['Club'] == 'New England Revolution']['Overall']

plt.figure(figsize=(5,9))

labels = ['FC Barcelona', 'Real Madrid', 'NE Revolution']
plt.boxplot([barcelona, madrid, revs], labels=labels)
plt.title('Professional Soccer Team Comparison')
plt.ylabel('Overall Ratings')

plt.show()
```

```python
## To change the style of each box, we can use a for loop
plt.figure(figsize=(5,9))

labels = ['FC Barcelona', 'Real Madrid', 'NE Revolution']
boxes = plt.boxplot([barcelona, madrid, revs], labels=labels)

for box in boxes['boxes']:
    box.set(color='#4286f4', linewidth=2)

plt.title('Professional Soccer Team Comparison')
plt.ylabel('Overall Ratings')

plt.show()
```
